```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# If not already loaded, make sure your data is in 'df'
# Example:
# df = pd.read_csv('your_file.csv')

# Step 1: Select key metrics
metrics = ['PE_Ratio', 'Dividend_yield_%', 'Qtr_Profit_%', 'Qtr_Sales_%', 'ROCE_%', 'ROE_%']
data = df[['Company_Name'] + metrics].copy()
data.set_index('Company_Name', inplace=True)

# Step 2: Invert PE ratio (lower is better, so inverse it)
data['PE_Ratio'] = 1 / data['PE_Ratio']

# Step 3: Handle infinite or NaN values (after inversion)
data = data.replace([np.inf, -np.inf], np.nan).dropna()

# Step 4: Normalize all metrics (0 to 1)
scaler = MinMaxScaler()
normalized_data = pd.DataFrame(scaler.fit_transform(data), columns=metrics, index=data.index)

# Step 5: Select top 3 companies by average score (optional)
normalized_data['Average_Score'] = normalized_data.mean(axis=1)
top_companies = normalized_data.sort_values(by='Average_Score', ascending=False).head(3)
normalized_data.drop(columns='Average_Score', inplace=True)

# Step 6: Create Radar Chart
labels = metrics
num_vars = len(labels)
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]  # close the circle

plt.figure(figsize=(8, 8))
for company in top_companies.index:
    values = normalized_data.loc[company].tolist()
    values += values[:1]
    plt.polar(angles, values, label=company, linewidth=2)
    plt.fill(angles, values, alpha=0.1)

plt.xticks(angles[:-1], labels)
plt.title("Top Financial Companies: Radar Chart of Financial Strength", size=14)
plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Choose key metrics
metrics = ['ROE_%', 'ROCE_%', 'Dividend_yield_%', 'Qtr_Profit_%', 'Qtr_Sales_%']

# Step 2: Calculate average score for each company
df['Overall_Score'] = df[metrics].mean(axis=1)

# Step 3: Sort by overall score (highest first)
df_sorted = df.sort_values(by='Overall_Score', ascending=False)

# Step 4: Plot
plt.figure(figsize=(12, 8))
sns.barplot(x='Overall_Score', y='Company_Name', data=df_sorted, palette='viridis')
plt.title("Overall Strength of Financial Companies (Average of Key Metrics)")
plt.xlabel("Overall Score")
plt.ylabel("Company Name")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Choose key metrics
metrics = ['ROE_%', 'ROCE_%', 'Dividend_yield_%', 'Qtr_Profit_%', 'Qtr_Sales_%']

# Step 2: Calculate average score
df['Overall_Score'] = df[metrics].mean(axis=1)

# Step 3: Select Top 10 companies
top10 = df.sort_values(by='Overall_Score', ascending=False).head(10)

# Step 4: Plot
plt.figure(figsize=(12, 6))
sns.barplot(x='Overall_Score', y='Company_Name', data=top10, palette='mako')
plt.title("Top 10 Financial Companies by Overall Strength")
plt.xlabel("Average Score (ROE, ROCE, Yield, Profit %, Sales %)")
plt.ylabel("Company Name")
plt.tight_layout()
plt.show()
```
